# Nigerian Agent Swarm Demo
This notebook demonstrates Task A (review generation) and Task B (recommendation) with Nigerian localization.

```python
import requests
import json

BASE_URL = "http://localhost:8000/api/v1"

def check_health():
    r = requests.get("http://localhost:8000/health")
    print(r.json())

check_health()
```

## Example 1: Task A — Lagos Hustler Review
Generate a Nigerian-pidgin review for a local food spot.

```python
payload = {
    "user_persona": {
        "user_id": "lagos_hustler_01",
        "demographics": "25yo, Lagos Island",
        "linguistic_register": "pidgin",
        "rating_bias": 0.3,
        "category_preferences": ["food", "transport"],
        "typical_review_length": "short",
        "emotional_triggers": ["NEPA", "traffic"]
    },
    "item_metadata": {
        "item_id": "rest_001",
        "name": "Mama Put Jollof Spot",
        "category": "food",
        "description": "Local buka serving smoky party jollof and grilled turkey",
        "attributes": {"cuisine": "Nigerian", "price_range": "₦", "location": "Yaba"}
    },
    "context": {
        "time_of_day": "afternoon",
        "day_of_week": "weekday",
        "mood_hint": "hungry after danfo struggle"
    }
}

r = requests.post(f"{BASE_URL}/task-a/generate-review", json=payload)
print(json.dumps(r.json(), indent=2))
```

## Example 3: Task B — Cross-Domain Recommendation
Recommend movies to a user with only food history (cold-start / cross-domain).

```python
payload3 = {
    "user_persona": {
        "user_id": "lagos_hustler_01",
        "demographics": "25yo, Lagos Island",
        "linguistic_register": "pidgin",
        "taste_vector": {"food": ["spicy", "cheap", "local"], "movies": ["Nollywood", "comedy"]},
        "cross_domain_signals": ["likes_local_culture", "budget_conscious"]
    },
    "context": {
        "conversation_history": [],
        "current_intent": "weekend movie night",
        "domain": "movies",
        "constraints": {"budget": "₦2000", "location": "Lagos"}
    },
    "top_k": 3
}

r = requests.post(f"{BASE_URL}/task-b/recommend", json=payload3)
print(json.dumps(r.json(), indent=2))
```
